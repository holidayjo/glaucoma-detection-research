```python
# Importing Necessary Libraries
from skimage import data
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import png
from skimage import io
from skimage import filters 
 
# Setting the plot size to 15,15
plt.figure(figsize=(15, 15))

img = io.imread("r1_Im069.png")
io.imshow(img)
plt.subplot(1, 2, 1)
 
# Displaying the sample image
plt.imshow(img)
 
# Converting RGB image to Monochrome
gray_img = rgb2gray(img)
plt.subplot(1, 2, 2)

# Displaying the sample image - Monochrome
# Format
plt.imshow(gray_img,  cmap="gray")

# Setting the plot size to 15,15
plt.figure(figsize=(15, 15))
 
for i in range(10):
   
  # Iterating different thresholds
  binarized_gray = (gray_img > i*0.1)*1
  plt.subplot(5,2,i+1)
   
  # Rounding of the threshold
  # value to 1 decimal point
  plt.title("Threshold: >"+str(round(i*0.1,1)))
   
  # Displaying the binarized image
  # of various thresholds
  plt.imshow(binarized_gray, cmap = 'gray')
   
plt.tight_layout()

# Setting the plot size to 15,15
plt.figure(figsize=(15, 15))

img2 = io.imread("r1_Im001.png")
io.imshow(img2)
plt.subplot(1, 2, 1)
 
# Displaying the sample image
plt.imshow(img2)
 
# Converting RGB image to Monochrome
gray_img2 = rgb2gray(img2)
plt.subplot(1, 2, 2)
 
# Displaying the sample image - Monochrome
# Format
plt.imshow(gray_img2, cmap="gray")

# Setting the plot size to 15,15
plt.figure(figsize=(15, 15))
 
for i in range(10):
   
  # Iterating different thresholds
  binarized_gray2 = (gray_img2 > i*0.1)*1
  plt.subplot(5,2,i+1)
   
  # Rounding of the threshold
  # value to 1 decimal point
  plt.title("Threshold: >"+str(round(i*0.1,1)))
   
  # Displaying the binarized image
  # of various thresholds
  plt.imshow(binarized_gray2, cmap = 'gray')
   
plt.tight_layout()
```

```python
# Importing Necessary Libraries
from skimage import data
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import png
from skimage import io
from skimage import filters
from skimage import morphology
from skimage import exposure

# Setting the plot size to 15,15
plt.figure(figsize=(15, 15))

img = io.imread("r1_Im069.png")
io.imshow(img)
plt.subplot(2, 2, 1)

# Displaying the sample image
plt.title("(1) Sample Diseased Eye")
plt.imshow(img)
 
# Converting RGB image to Monochrome
gray_img = rgb2gray(img)
plt.subplot(2, 2, 2)

# Displaying the sample image - after monochrome filter
plt.title("(2) Sample after monochrome")
plt.imshow(gray_img, cmap="gray")

clahe_img = exposure.equalize_adapthist(gray_img)
plt.subplot(2, 2, 3)

# Displaying sample image - after Contrast Limited Adaptive Histogram Equalization (CLAHE)
plt.title("(3) Sample after CLAHE algorithm")
plt.imshow(clahe_img, cmap="gray")

plt.tight_layout()
```
