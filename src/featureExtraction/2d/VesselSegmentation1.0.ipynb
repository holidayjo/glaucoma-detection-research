```python
from skimage import data
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import png
from skimage import io
from skimage import filters
from skimage import morphology
from skimage import exposure
from skimage import restoration
import numpy as np
import cv2

plt.figure(figsize=(15, 15))

img = io.imread("r1_Im069.png")
io.imshow(img)
plt.subplot(4, 2, 1)

# Display sample image
plt.title("(1) Sample Diseased Eye")
plt.imshow(img)
 
# Convert RGB image to monochrome
gray_img = rgb2gray(img)
plt.subplot(4, 2, 2)
plt.title("(2) Monochrome Conversion")
plt.imshow(gray_img, cmap="gray")

# Equalize contrast with CLAHE algorithm - Contrast Limited Adaptive Histogram Equalization
# (Uses histrograms computed over different tile regions to enhance local details irregardless of brightness)
clahe_img = exposure.equalize_adapthist(gray_img)
plt.subplot(4, 2, 3)
plt.title("(3) Contrast Equalization w/ CLAHE")
plt.imshow(clahe_img, cmap="gray")

# Denoise image with the median blur filter
denoised_img = cv2.medianBlur(np.float32(clahe_img), 3)
plt.subplot(4, 2, 4)
plt.title("(4) Median Blur Filter w/ Kernel Size 3")
plt.imshow(denoised_img, cmap="gray")

# Find way to improve separation of blood vessels before producing binarized image...

# Iterate over threshold values to produce binarized image
# plt.figure(figsize=(30, 30))
# threshold = 0.00
# increment = 0.05
# for i in range(20):
#     binarized_gray = (gray_img > threshold) * 1
#     plt.subplot(10, 2, i+1)
#     plt.title("Threshold: > " + str(round(threshold, 2)))
#     plt.imshow(binarized_gray, cmap = 'gray')
#     threshold += increment

# Compare various thresholding methods
fig, ax = filters.try_all_threshold(denoised_img, figsize=(10, 8), verbose=False)
plt.show()

plt.tight_layout()
```
