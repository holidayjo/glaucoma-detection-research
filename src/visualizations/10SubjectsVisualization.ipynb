```python
import matplotlib.pyplot as plt
from skimage import io

imagesDiseased = []
imagesHealthy = []
for i in range(1, 11):
    imagesDiseased.append("glaucoma" + str(i) + ".png")
       
for i in range(1, 11):   
    imagesHealthy.append("healthy" + str(i) + ".png")
    
imagesDiseasedMasks = []
imagesHealthyMasks = []
for i in range(1, 11):
    imagesDiseasedMasks.append("./masks/glaucoma" + str(i) + "mask.png")
       
for i in range(1, 11):   
    imagesHealthyMasks.append("./masks/healthy" + str(i) + "mask.png")
    
imagesDiseasedMeshFront = []
imagesHealthyMeshFront = []
for i in range(1, 11):
    imagesDiseasedMeshFront.append("./meshScreenshots/front/glaucoma" + str(i) + "front.png")
       
for i in range(1, 11):   
    imagesHealthyMeshFront.append("./meshScreenshots/front/healthy" + str(i) + "front.png")
    
imagesDiseasedMeshRight = []
imagesHealthyMeshRight = []
for i in range(1, 11):
    imagesDiseasedMeshRight.append("./meshScreenshots/right/glaucoma" + str(i) + "right.png")
       
for i in range(1, 11):   
    imagesHealthyMeshRight.append("./meshScreenshots/right/healthy" + str(i) + "right.png")
    
imagesDiseasedMeshLeft = []
imagesHealthyMeshLeft = []
for i in range(1, 11):
    imagesDiseasedMeshLeft.append("./meshScreenshots/left/glaucoma" + str(i) + "left.png")
       
for i in range(1, 11):   
    imagesHealthyMeshLeft.append("./meshScreenshots/left/healthy" + str(i) + "left.png")

plt.figure(figsize=(10, 10))

i = 1
for imgSource in imagesHealthyMeshLeft:
    img = io.imread(imgSource)
    plt.subplot(10, 10, i)
    plt.title('left ' + str(i))
    ax = plt.gca()
    ax.xaxis.set_tick_params(labelbottom=False)
    ax.yaxis.set_tick_params(labelleft=False)
    ax.set_xticks([])
    ax.set_yticks([])
    plt.imshow(img)
    i += 1
    
plt.tight_layout()
```
